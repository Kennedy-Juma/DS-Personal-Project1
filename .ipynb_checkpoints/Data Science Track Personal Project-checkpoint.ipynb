# Developing a model to train and test the data

```python
df_HR.info()
```

```python
Emp_Attrition=['Attrition']
Emp_BusinessTravel=['BusinessTravel']
Emp_Department=['Department']
Emp_EducationField=['EducationField']
Emp_Gender=['Gender ']
Emp_Role=['JobRole']
Emp_MaritalStatus=['MaritalStatus']
Emp_Over18 =['Over18 ']
Emp_OverTime=['OverTime']
```

```python
final_data=pd.get_dummies(df_HR,['Emp_Attrition','Emp_BusinessTravel','Emp_Department','Emp_EducationField',
                         'Emp_Gender','Emp_Role','Emp_MaritalStatus','Emp_Over18','Emp_OverTime'],drop_first=True)
```

```python
final_data.info()
```

```python
#Setting target and predictor variables
col=list(final_data)
predictors=final_data[col[1:46]]
target=final_data['PerformanceRating']
```

```python
#creating thre training and testing data
from sklearn.model_selection import train_test_split
```

```python
predictors_train,predictors_test,target_train,target_test=train_test_split(predictors,target,test_size=0.25,random_state=0)
```

```python
#Train the RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=500)
clf.fit(predictors_train,target_train)
```

```python
#Applying the classifier to test the data
target_pred=clf.predict(predictors_test)
```

```python
#Calculating the score
clf.score(predictors_test,target_test)
```

```python
#Generating a confusion matrix inorder to predict actual and predicted values
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(target_test,target_pred)
print(cm)
print(classification_report(target_test,target_pred))
```

```python
#Creating feature importance
feature_importances = pd.DataFrame(data=clf.feature_importances_*100,index=predictors_train.columns,
                                    columns=['importance']).sort_values('importance', ascending=False)
```
